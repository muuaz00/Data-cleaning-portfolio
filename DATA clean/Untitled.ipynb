```python
import numpy as np
df
```

```python
# Replace NaN with '-'
df = df.fillna('-')
df
```

```python
import re
def remove_special_chars(text):
    if isinstance(text, str):  # Only process strings
        return re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text
df['Tour title	'] = df['Tour title'].map(remove_special_chars)
df = df.drop('Ref.', axis=1)
df
```

```python
def clean_brackets(value):
    if isinstance(value, str):  # Only process strings
        match = re.match(r'^(\d+)\[.*\]$', value)  # Match pattern like '2[3]'
        if match:
            return match.group(1)  # Return number before brackets
    return value

# Apply to all columns
df = df.apply(lambda col: col.map(clean_brackets))
df
```

```python
df.to_csv('output.csv', index=False)
```

```python
def clean_years(value):
    if isinstance(value, str) and '-' in value:  # Check for strings with hyphen
        return value.replace('-', ' ')  # Replace hyphen with space
    return value

# Apply cleaning to 'years' column
df['Year(s)'] = df['Year(s)'].map(clean_years)
df
```

```python
df.to_csv('output.csv', index=False)
```

```python
def split_years(value):
    if isinstance(value, str) and '–' in value:  # Check for en-dash
        start, end = value.split('–')
        return pd.Series([start.strip(), end.strip()])
    elif isinstance(value, str):  # Handle single years
        return pd.Series([value.strip(), None])
    return pd.Series([None, None])

# Apply splitting to 'Year(s)' column
df[['Start_Year', 'End_Year']] = df['Year(s)'].apply(split_years)

# Drop the original 'Year(s)' column (optional, comment out to keep it)
df = df.drop('Year(s)', axis=1)
df
```

```python
df.to_csv('my_file_split_years.csv', index=False, encoding='utf-8-sig')
```
